We can Fit a linear regression model using the Cases - newly reported in last 24 hours feature and caculate the Deaths - newly reported in last 24 hours

```python
# Import Linear Regression library
%conda install sklearn
from sklearn.linear_model import LinearRegression 
```

```python
americas_df.dtypes
```

```python
x = americas_df[['Daily_Cases']]
y = americas_df['Weekly_Deaths'] 
lm = LinearRegression() 
lm.fit(x,y)
lm
```

What is the value of the intercept ?

```python
lm.intercept_
```

What is the value of the Slope ?

```python
lm.coef_
```

<h2>Model performance evaluation metrics</h2>

Model Score R Squared

```python
lm.score(x, y)
```

Correlation

mean_squared_error % sqrt from the module metrics

```python
#import libraries
from sklearn.metrics import mean_squared_error #for calculating MSE
from math import sqrt #for calculating RSME
```

```python
#calculating MSE
y=lm.predict(x)
mse = mean_squared_error(americas_df['Daily_Deaths'], y)
print('Mean Squared Error:', mse)
```

```python
#calculating RMSE
#since; mse=sqrt(mean_squared_error(americas_df['Daily_Deaths'], y))
print('Root Mean Squared Error:',sqrt(mse))
```

```python
americas_df[['Daily_Cases', 'Weekly_Deaths']].corr()
```

Prediction

```python
Yhat=lm.predict(x)
Yhat[0:5]
```

<h1>Multiple Linear Regression</h1>

Predicting Weely Deaths in America by looking at three variables;
1. Daily_Cases
2. Weekly_Cases 
3. Total_Cases

```python
# assign the variables identified above; in a new dataset,z
z = americas_df[['Daily_Cases','Weekly_Cases','Total_Cases']]
#Fitting the linear model using the three above variables.
lm.fit(z, americas_df['Weekly_Deaths'])
```

the value of the intercept, c

```python
lm.intercept_
```

value of the slope; b1,b2,b3

```python
lm.coef_
```

Model formular

$$Daily_Deaths = 2.00 * Daily_Cases - 2.47 * Weekly_Cases - 1.45 * Total_Cases -1.7$$
As of July 9th 2021 data by WHO. This model keeps changing as data is being updated daily

correlation

```python
americas_df[['Daily_Cases','Weekly_Cases','Total_Cases']].corr()
```

R-squared

```python
print('The R-squared is:',lm.score(z, americas_df['Weekly_Deaths']))
```

Mean Squared Error; MSE

```python
y_predict_multifit = lm.predict(z)
#Now we can compare the predicted results with the actual results and print
mse=mean_squared_error(americas_df['Weekly_Deaths'], y_predict_multifit)
print('Mean Squared Error:',mse)
```

mean_squared_error

```python
print('Root Mean Squared Error:',sqrt(mse))
```
